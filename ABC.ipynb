```
import pandas as pd
```

```
df = pd.read_csv('orders_and_shipments.csv')
```

```


# Sample data from the context provided

# Calculate total value for each item
df['TotalValue'] = df['Order Quantity'] * df[' Profit ']

# Sort items by total value in descending order
df = df.sort_values(by='TotalValue', ascending=False)

# Calculate cumulative sum and percentage of total value
df['CumulativeValue'] = df['TotalValue'].cumsum()
total_value = df['TotalValue'].sum()
df['CumulativePercentage'] = (df['CumulativeValue'] / total_value) * 100

# Classify items into A, B, and C categories
def classify_abc(row):
    if row['CumulativePercentage'] <= 80:
        return 'A'
    elif row['CumulativePercentage'] <= 95:
        return 'B'
    else:
        return 'C'

df['Category'] = df.apply(classify_abc, axis=1)

# Display the result
print(df[['Category', 'TotalValue', 'CumulativePercentage']])
```
