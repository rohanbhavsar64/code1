```
import pandas as pd
```

```
df = pd.read_csv('orders_and_shipments.csv')
```

```
df
```

```
df['Unit Price'] = df[' Gross Sales '] /df['Order Quantity']
df
```

```
df['TotalValue'] = df['Order Quantity']
df['TotalValue1'] = df['Unit Price']
df
df = df.sort_values(by='TotalValue', ascending=False)
df['CumulativeValue'] = df['TotalValue'].cumsum()
total_value = df['TotalValue'].sum()
df['CumulativePercentage'] = (df['CumulativeValue'] / total_value) * 100
df = df.sort_values(by='TotalValue1', ascending=False)
df['CumulativeValue1'] = df['TotalValue1'].cumsum()
total_value = df['TotalValue1'].sum()
df['CumulativePercentage1'] = (df['CumulativeValue1'] / total_value) * 100
def classify_abc(row):
    if (row['CumulativePercentage1'] < 80) and (row['CumulativePercentage'] < 20) :
        return 'A'
    elif ((row['CumulativePercentage1'] >80) and (row['CumulativePercentage1'] < 95) and (row['CumulativePercentage'] >20) and (row['CumulativePercentage'] < 50) ):
        return 'B'
    else:
        return 'C'

df['Category'] = df.apply(classify_abc, axis=1)
print(df[['Category', 'TotalValue', 'CumulativePercentage']])
```
